# Stat 200B Data Visualization Intro
## How can we use visualization to understand the data?  

### First, let's load some packages to give us some computing power.

```python
import statsmodels.api as sm
import sklearn
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sb

#Allow Multiple Outputs
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"


```

### Now, let's read the file and see what it looks like

```python
df=pd.read_csv(r'Stat200_Data.csv')
df.head()
```

```python
#How does Pandas interpret the data? 
df.columns
df.info()
```

```python
sb.countplot(df['MaritalStatus'])
```

```python
 sb.distplot(df['Income'], bins=10)
```

```python
#sb.countplot(data['AgeHeadHousehold'])
sb.distplot(df['AgeHeadHousehold'])
```

```python
sb.countplot(df['AgeHeadHousehold'])
```

```python
sb.scatterplot(x=df['AgeHeadHousehold'], y=df['Income'])

```

### Doesn't seem to be a large correlation between age and income (which is usually typical). What if we plot mroe stuff?

```python
data_set=df[['Income','AgeHeadHousehold', 'Food', 'Meat', 'Bakery', 'Fruits']]

sb.pairplot(data=data_set)

#sb.distplot??
```

#And what if we separate based on marital status? 

```python
sb.set_palette('deep')

data_set=df[['MaritalStatus','Income', 'AgeHeadHousehold','Food', 'Meat', 'Bakery', 'Fruits']]

sb.pairplot(data=data_set, hue='MaritalStatus')



#sb.distplot??
```
